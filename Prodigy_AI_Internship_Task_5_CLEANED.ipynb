```python
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Function to load and preprocess images
def load_image(path_to_img, max_dim=512):
    img = Image.open(path_to_img)
    img = img.convert('RGB')
    img.thumbnail((max_dim, max_dim))
    img = np.array(img)
    img = img.astype(np.float32)[np.newaxis, ...] / 255.0
    return img

# Upload your content and style images
from google.colab import files
uploaded = files.upload()

# Provide your image file names here (replace with your uploaded filenames)
content_image_path = "/content/Arjunudu.jpeg"
style_image_path = "/content/ChatGPT Image May 13, 2025, 10_08_47 PM.png"

# Load images
content_image = load_image(content_image_path)
style_image = load_image(style_image_path)

# Load the style transfer model from TensorFlow Hub
style_transfer_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')

# Apply style transfer
stylized_image = style_transfer_model(tf.constant(content_image), tf.constant(style_image))[0]

# Show result
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Content Image')
plt.imshow(content_image[0])
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Style Image')
plt.imshow(style_image[0])
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Stylized Image')
plt.imshow(stylized_image[0])
plt.axis('off')

plt.show()
```
